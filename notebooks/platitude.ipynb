```python
from platitude.cliche.bm25 import Bm25
from platitude.cliche.table import merge_columns
import json
from tqdm.autonotebook import tqdm
from IPython.display import display
from tabulate import tabulate
from tempfile import TemporaryDirectory
from pathlib import Path
import ipywidgets as widgets
```

```python
tmp = TemporaryDirectory()

def lines(president: str):
    global oups
    for i, line in tqdm(enumerate(f.readlines()), total=total, desc=f"Index {president} speeches"):
        doc = json.loads(line)
        txt = doc['full_text']
        if txt is not None:
            yield txt
        else:
            oups += 1

tops = dict()

for president in ['chirac', 'sarkozy', 'hollande', 'macron']:
    path = Path(tmp.name) / president
    path.mkdir()
    idx = Bm25(path,4, 6)
    f = open(f'../{president}.jsonl', 'r')
    total = len(f.readlines())
    f.seek(0)
    oups = 0

    print("oups", oups)
    idx.index(lines(president))
    idx.flush()

    assert len(idx.documents) + oups == total, len(idx.documents)
    print("vocab", len(idx.vocabulary))

    top = idx.top(20)
    tops[president] = top
    print("compression ratio:", idx.documents.data.compression_ratio())

tmp.cleanup()
```

```python
headers = [" "] * len(tops) * 2
for i, prez in enumerate(tops.keys()):
    headers[i*2] = prez
display(tabulate(merge_columns(*merge_columns(tops.values())), headers, tablefmt='html'))
```
