```python
from platitude.cliche.bm25 import Bm25
import json
from tqdm import tqdm
import re
from IPython.display import display
from tabulate import tabulate
```

```python
idx = Bm25()

f = open('../chirac.jsonl', 'r')
total = len(f.readlines())
f.seek(0)

for line in tqdm(f.readlines(), total=total, desc="Index docs"):
    doc = json.loads(line)
    idx.index([doc['full_text']])

assert len(idx.documents) == total, len(idx.documents)

idx.compute()

display(tabulate(idx.top(), tablefmt='html'))
```
