```python
from platitude.cliche.bm25 import Bm25
import json
from tqdm import tqdm
from IPython.display import display
from tabulate import tabulate
from tempfile import TemporaryDirectory
from pathlib import Path
```

```python
tmp = TemporaryDirectory()
print(tmp.name)
idx = Bm25(Path(tmp.name),4, 6)

f = open('../chirac1000.jsonl', 'r')
total = len(f.readlines())
f.seek(0)
oups = 0

def lines():
    global oups
    for i, line in tqdm(enumerate(f.readlines()), total=total, desc="Index docs"):
        doc = json.loads(line)
        txt = doc['full_text']
        if txt is not None:
            yield txt
        else:
            oups += 1
    print("oups", oups)
idx.index(lines())
idx.flush()

assert len(idx.documents) + oups == total, len(idx.documents)
print("vocab", len(idx.vocabulary))

top100 = sorted(idx.vocab_stats(), key=lambda x: x[1], reverse=True)[:100]
display(tabulate(top100, tablefmt='html'))

tmp.cleanup()
```
