```python
from platitude.cliche.bm25 import Bm25
import json
from tqdm.autonotebook import tqdm
from IPython.display import display
from tabulate import tabulate
from tempfile import TemporaryDirectory
from pathlib import Path
import ipywidgets as widgets
```

```python
president = widgets.Select(
    options=['chirac', 'sarkozy', 'hollande', 'macron'],
    value='chirac',
    description='President:',
    disabled=False
)
display(president)
```

```python
tmp = TemporaryDirectory()
idx = Bm25(Path(tmp.name),4, 6)

f = open(f'../{president.value}.jsonl', 'r')
total = len(f.readlines())
f.seek(0)
oups = 0

def lines():
    global oups
    for i, line in tqdm(enumerate(f.readlines()), total=total, desc=f"Index {president.value} speeches"):
        doc = json.loads(line)
        txt = doc['full_text']
        if txt is not None:
            yield txt
        else:
            oups += 1
    print("oups", oups)
idx.index(lines())
idx.flush()

assert len(idx.documents) + oups == total, len(idx.documents)
print("vocab", len(idx.vocabulary))

top100 = sorted(idx.vocab_stats(), key=lambda x: x[1], reverse=True)[:100]
display(tabulate(top100, tablefmt='html'))
print("compression ratio:", idx.documents.data.compression_ratio())

tmp.cleanup()
```
